```python
from collections import defaultdict

def compute_elo(df, k=32, init_elo=1000, surface_col="Surface"):
    # returns df with winner_elo_pre, loser_elo_pre, elo_diff and also updates internals
    df = df.copy()
    # initialize
    elo = defaultdict(lambda: init_elo)
    elo_surface = defaultdict(lambda: defaultdict(lambda: init_elo))
    w_elo_pre, l_elo_pre, w_elo_surf_pre, l_elo_surf_pre = [], [], [], []

    for _, row in df.iterrows():
        w, l = row["Winner"], row["Loser"]
        surf = row.get(surface_col, None)

        # pre-match
        w_pre = elo[w]
        l_pre = elo[l]
        w_surf_pre = elo_surface[w][surf]
        l_surf_pre = elo_surface[l][surf]

        w_elo_pre.append(w_pre)
        l_elo_pre.append(l_pre)
        w_elo_surf_pre.append(w_surf_pre)
        l_elo_surf_pre.append(l_surf_pre)

        # expected
        exp_w = 1 / (1 + 10 ** ((l_pre - w_pre) / 400))
        exp_l = 1 - exp_w

        # update global
        elo[w] += k * (1 - exp_w)
        elo[l] += k * (0 - exp_l)

        # update surface
        exp_w_s = 1 / (1 + 10 ** ((l_surf_pre - w_surf_pre) / 400))
        elo_surface[w][surf] += k * (1 - exp_w_s)
        elo_surface[l][surf] += k * (0 - (1 - exp_w_s))

    df["winner_elo_pre"] = w_elo_pre
    df["loser_elo_pre"] = l_elo_pre
    df["elo_diff"] = df["winner_elo_pre"] - df["loser_elo_pre"]
    df["winner_elo_surf_pre"] = w_elo_surf_pre
    df["loser_elo_surf_pre"] = l_elo_surf_pre
    df["elo_surf_diff"] = df["winner_elo_surf_pre"] - df["loser_elo_surf_pre"]
    return df

df = compute_elo(df)
```

```python
def add_h2h_feature(df):
    df = df.copy()
    h2h = defaultdict(lambda: defaultdict(int))
    h2h_pre = []

    for _, row in df.iterrows():
        w, l = row["Winner"], row["Loser"]

        # record value BEFORE match
        h2h_pre.append(h2h[w][l] - h2h[l][w])

        # update AFTER match
        h2h[w][l] += 1

    df["h2h_pre"] = h2h_pre
    return df

df = add_h2h_feature(df)
```

```python
from collections import deque
def add_recent_form(df, N=10):
    df = df.copy()
    recent_form = defaultdict(lambda: deque(maxlen=N))
    form_pre = []

    for _, row in df.iterrows():
        w, l = row["Winner"], row["Loser"]

        # wins so far before match
        w_wins = sum(recent_form[w])
        l_wins = sum(recent_form[l])
        form_pre.append(w_wins - l_wins)

        # update after match
        recent_form[w].append(1)
        recent_form[l].append(0)

    df["recent_form_diff"] = form_pre
    return df

df = add_recent_form(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()

for header in list(df.columns):
    if (df[header].dtype == object):
        df[header] = ord_enc.fit_transform(df[[header]])

df.info()
```
